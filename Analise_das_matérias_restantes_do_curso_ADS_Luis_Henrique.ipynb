```python
import pandas as pd
```

```python
colunms = ['Ano Período',	'Período', 'COD Componente Curricular', 'Componente Curricular',	'Tipo',	\
           'Carga Horária',	'Créditos',	'Nota', 'Frequência',	'Situação']
```

```python
data = pd.read_html('/content/Q-Acadêmico Web para IFPI - Histórico Escolar.html',
                    attrs = {'id': "materias"})
```

```python
df = data[0]
```

```python
df.head()
```

```python
df.columns = colunms
```

```python
df.head()
```

```python
df = df.drop([0])
```

```python
data_mask = df['Situação'] != 'Aprovado'
```

```python
data_mask2 = df['Situação'] == 'Aprovado'
```

```python
df_rep = df[data_mask]
df_rep
```

```python
df_apro = df[data_mask2]
df_apro
```

```python
aprov = list(df_apro['Componente Curricular'])
aprov
```

```python
mask = [not ma in aprov for ma in list(df_rep['Componente Curricular'])]
```

```python
materias_restantes = df_rep[mask]
```

```python
materias_restantes
```

```python
disciplinas_res = set(list(materias_restantes['Componente Curricular']))
disciplinas_res
```
